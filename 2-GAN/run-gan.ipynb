# GAN(Generative Adverserial Netork)

```python
IMG_SHAPE= (28,28,1)
NOISE_DIM = 100  
```

```python
from data_generator import *

x_train = get_fashion_mnist()
x_train.shape
```

```python
from build_models import *

builder = BuildModels(IMG_SHAPE, NOISE_DIM)
gene = builder.build_generator()
disc = builder.build_discriminator()
```

```python
from gan import *

model = GAN(gene, disc, IMG_SHAPE, NOISE_DIM)
model.compile()
model.gan.summary()
```

```python
history = model.train(x_train, epochs=5)
```
