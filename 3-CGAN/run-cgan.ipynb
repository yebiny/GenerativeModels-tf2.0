# CGAN (conditional generative adversarial network)

```python
IMG_SHAPE = (28,28,1)
NOISE_DIM = 100
LABEL_DIM = 10
```

```python
from data_generator import *

x_data, y_data = get_fashion_mnist()
x_data.shape, y_data.shape
```

```python
from build_models import *

builder = BuildModels(IMG_SHAPE, NOISE_DIM, LABEL_DIM)
gene = builder.build_generator()
disc = builder.build_discriminator()
```

```python
from cgan import *

model = CGAN(gene, disc, IMG_SHAPE, NOISE_DIM, LABEL_DIM)
model.compile()
plot_model(model.cgan)
```

```python
history = model.train( x_data, y_data
                     , epochs=5
                     , batch_size=16
                     )
```
